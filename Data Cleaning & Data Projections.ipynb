# Data Cleaning

#### Economic Activity by Age 2011 and 2021

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_2011_tot = pd.read_csv("Raw Data/Total Economic Activity by Age 2011.csv")
df_2011_emp = pd.read_csv("Raw Data/Employment Economic Activity by Age 2011.csv")
```

```python
df_2011_tot
```

```python
df_2011_emp
```

```python
df_2011_emp.columns
```

```python
df_2011_melt_tot = pd.melt(df_2011_tot, id_vars=['Economic Activity','Year', 'Middle layer Super Output Areas Code',
       'Middle layer Super Output Areas'], value_vars=['Total', 'Age 16 to 24',
       'Age 25 to 34', 'Age 35 to 49', 'Age 50 to 64', 'Age 65 and over'])
df_2011_melt_emp = pd.melt(df_2011_emp, id_vars=['Economic Activity','Year', 'Middle layer Super Output Areas Code',
       'Middle layer Super Output Areas'], value_vars=['Total', 'Age 16 to 24',
       'Age 25 to 34', 'Age 35 to 49', 'Age 50 to 64', 'Age 65 and over'])
```

```python
df_2011_melt_emp
```

```python
df_concat = pd.concat([df_2011_melt_tot,df_2011_melt_emp])
df_concat.to_csv("Raw Data/Combined Economic Activity by Age 2011.csv", index=False)
```

```python
df_2021 = pd.read_csv("Raw Data/Economic Activity by Age 2021 RAW.csv")
```

```python
df_2021
```

```python
df_2021_drop = df_2021.drop(df_2021.loc[df_2021['Age']=='Aged 15 years and under'].index)
df_2021_drop
```

```python
df_2021_drop['Age'] = df_2021_drop['Age'].replace('Aged 16 to 24 years','Age 16 to 24')
df_2021_drop['Age'] = df_2021_drop['Age'].replace('Aged 16 to 24 years','Age 16 to 24')
df_2021_drop['Age'] = df_2021_drop['Age'].replace('Aged 25 to 34 years','Age 25 to 34')
df_2021_drop['Age'] = df_2021_drop['Age'].replace('Aged 35 to 49 years','Age 35 to 49')
df_2021_drop['Age'] = df_2021_drop['Age'].replace('Aged 50 to 64 years','Age 50 to 64')
df_2021_drop['Age'] = df_2021_drop['Age'].replace('Aged 65 years and over','Age 65 and over')
```

```python
df_2021_drop
```

```python
df_2011_clean = pd.read_csv("Raw Data/Combined Economic Activity by Age 2011.csv")
```

```python
df_merge = df_2011_clean.merge(df_2021_drop, how = "inner", on = "Middle layer Super Output Areas Code")
```

```python
df_merge
```

```python
df_2011 = df_merge.iloc[:, :6]
df_2021 = pd.concat([df_merge.iloc[:, 6], df_merge.iloc[:, 1],df_merge.iloc[:, 2], df_merge.iloc[:, 8:]], axis=1)
df_2011 = df_2011.rename(columns={"Year_x":"Year", "Middle layer Super Output Areas_x": "Middle layer Super Output Areas", 'Age_x': 'Age', 'Economic Activity_x': 'Economic Activity', 'Observation_x': 'Observation'})
df_2021 = df_2021.rename(columns={"Year_y":"Year", "Middle layer Super Output Areas_x": "Middle layer Super Output Areas", 'Age_y': 'Age', 'Economic Activity_y': 'Economic Activity', 'Observation_y': 'Observation'})
```

```python
df_clean_merge = pd.concat([df_2011, df_2021])
```

```python
df_clean_merge
```

```python
df_clean_merge.to_csv("Cleaned Data/Economic Activity by Age 2011 and 2021 FINAL.csv", index=False)
```

# Data Projections

#### Data Projection 1: TSNE

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cleaned Data/Economic Activity by Highest Level of Qualification 2021 FINAL.csv")
```

```python
df_2_pivot = pd.pivot_table(df, values='Observation', index=['Middle layer Super Output Areas','Highest level of qualification (7 categories)'],columns=['Sex (2 categories)','Age (6 categories)','Economic activity status (4 categories)'], aggfunc="sum")
```

```python
df_2_pivot
```

```python
df_2_pivot.to_csv("Raw Data/Economic Activity by Highest Level of Qualification 2021 GROUPED RAW.csv",index=True)
df_2_pivot = pd.read_csv("Raw Data/Economic Activity by Highest Level of Qualification 2021 GROUPED RAW.csv")
df_2_pivot
```

```python
df_trans = df_2_pivot.iloc[3:,3:]
```

```python
tsne = TSNE(random_state=13)
transformed_val = tsne.fit_transform(df_trans)
fig, ax = plt.subplots()
plt.scatter(transformed_val[:,0],transformed_val[:,1])
```

```python
df_concat = pd.concat([df_2_pivot.iloc[3:,:2].reset_index(),pd.DataFrame(transformed_val)], axis=1)
```

```python
df_final = df_concat.drop(columns="index")
```

```python
df_final = df_final.rename(columns={'Sex (2 categories)':'Middle layer Super Output Areas','Unnamed: 1':'Highest level of qualification (7 categories)'})
df_final
```

```python
df_final.to_csv("Cleaned Data/Economic Activity by Highest Level of Qualification 2021 GROUPED FINAL.csv",index=False)
```

### Data Projection 2: UMAP

```python
import umap
df = pd.read_csv("Cleaned Data/Highest Level of Qualification by Working Conditions 2021 FINAL.csv")
```

```python
df_2_pivot = pd.pivot_table(df, values='Observation', index=['Middle layer Super Output Areas','Highest level of qualification (7 categories)'],columns=['Sex (2 categories)','Hours worked (5 categories)','Distance travelled to work (5 categories)'], aggfunc="sum")
```

```python
df_2_pivot
```

```python
df_2_pivot.to_csv("Raw Data/Highest Level of Qualification by Working Conditions 2021 GROUPED RAW.csv",index=True)
df_2_pivot = pd.read_csv("Raw Data/Highest Level of Qualification by Working Conditions 2021 GROUPED RAW.csv")
df_2_pivot
```

```python
df_trans = df_2_pivot.iloc[3:,3:]
reducer = umap.UMAP(random_state=13)
embedding = reducer.fit_transform(df_trans)
fig,ax = plt.subplots()
plt.scatter(embedding[:, 0], embedding[:, 1])
```

```python
df_concat = pd.concat([df_2_pivot.iloc[3:,:2].reset_index(),pd.DataFrame(embedding).reset_index()], axis=1)
```

```python
df_concat = df_concat.drop(columns="index")
```

```python
df_final = df_concat.rename(columns={'Sex (2 categories)':'Middle layer Super Output Areas','Unnamed: 1':'Highest level of qualification (7 categories)'})
df_final
```

```python
df_final.to_csv("Cleaned Data/Highest Level of Qualification by Working Conditions 2021 GROUPED FINAL.csv",index=False)
```
